# 🧠 Single Agent Pipeline Project

## Problem Statement
Build a **Single-Agent Smart Assistant** that:
- Understands user queries
- Routes tasks based on intent
- Uses tools when required
- Returns structured JSON output

### The agent should handle:
- Math queries → Calculator Tool
- Keyword extraction → Keyword Tool
- General queries → Direct response

---
### 🛠️ What You Need to Implement
- Agent logic
- Conditional routing
- Tool integration
- Basic error handling

### 🚀 Bonus
- Improve routing
- Add logging
- Add more tools


````python
# 🛠️ TOOL 1: Calculator (class-based)

class CalculatorTool:
    """Evaluates a mathematical expression passed in as a string."""

    def run(self, expression: str) -> str:
        try:
            output = eval(expression)
            return str(output)
        except Exception:
            return "Error in calculation"
````

````python
# 🛠️ TOOL 2: Keyword Extractor (class-based, regex driven)

import re

class KeywordTool:
    """Pulls out the most notable words from a chunk of text."""

    def run(self, text: str, limit: int = 5) -> list:
        try:
            tokens = re.findall(r"[A-Za-z]+", text)
            significant = sorted(set(token.lower() for token in tokens if len(token) > 4))
            return significant[:limit]
        except Exception:
            return []
````

## 🤖 Implement Agent Logic Below

👉 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

````python
# 🤖 AGENT (IMPLEMENTED as a class with dispatch-table routing)

class SmartAssistant:
    def __init__(self):
        self.calculator = CalculatorTool()
        self.keyword_tool = KeywordTool()
        # Order matters: first matching trigger wins
        self.routes = [
            ("calculate", self._handle_math),
            ("keywords", self._handle_keywords),
        ]

    def _handle_math(self, query: str, query_lower: str) -> dict:
        expression = query_lower.replace("calculate", "", 1).strip()

        if not expression:
            return {"type": "error", "result": "No expression found to calculate"}

        calc_result = self.calculator.run(expression)

        if calc_result == "Error in calculation":
            return {"type": "error", "result": calc_result}

        return {"type": "calculation", "result": calc_result}

    def _handle_keywords(self, query: str, query_lower: str) -> dict:
        keywords = self.keyword_tool.run(query)
        return {"type": "keywords", "result": keywords}

    def _handle_general(self, query: str, query_lower: str) -> dict:
        return {"type": "general", "result": f"This is a general query. You asked: '{query}'"}

    def respond(self, query: str) -> dict:
        query_lower = query.lower()
        try:
            for trigger, handler in self.routes:
                if trigger in query_lower:
                    return handler(query, query_lower)
            return self._handle_general(query, query_lower)
        except Exception as e:
            return {"type": "error", "result": f"Unexpected error: {str(e)}"}


assistant = SmartAssistant()
````

## 📦 Expected Output Format

```
{
  "type": "calculation / keywords / general / error",
  "result": ...
}
```
